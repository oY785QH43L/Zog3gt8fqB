## Initial model (Transact-SQL)

Execute all requirements and measure time needed to execute the Stored procedure or UDF and measure the time needed for every execution.
Repeat the measurements for all requirements for n times. Store the measurements as dataframe and save it as a CSV file.

```python
it_count = 30
```

Admin

```python
import pandas as pd
from admin.admin_test_data import AdminTestData
import admin.requirements_sql
import importlib
import time

importlib.reload(admin.requirements_sql)
from admin.requirements_sql import mapping_dictionary

config_data = AdminTestData()


def get_admin_measurements(iteration_count: int) -> pd.DataFrame:
    """Retrieves the operation measurements in seconds for the admin requirements.

    Args:
        iteration_count (int): The iteration count.

    Raises:
        ValueError: Is thrown if the iteration count is not positive.

    Returns:
        pd.DataFrame: The result dataframe.
    """
    if iteration_count <= 0:
        raise ValueError("iteration_count must be positive!")
    keys_lists = [("A" + str(e), []) for e in list(range(1, 10))]
    final_dict = dict(keys_lists)
    execution_order = [1, 2, 4, 5, 6, 7, 8, 9, 3]

    for i in range(iteration_count):
        for el in execution_order:
            key = "A" + str(el)
            elapsedTime = mapping_dictionary[key]()
            final_dict[key].append(elapsedTime)
            time.sleep(1)

    return pd.DataFrame(final_dict)


# Get the measurements and store them in a CSV
admin_frame = get_admin_measurements(it_count)
admin_frame.to_csv(config_data.admin_sql_measurements_path, index=False)
```

## Plot performance for requirements

Admin

```python
import pandas as pd
import matplotlib.pyplot as plt


def plot_admin_requirements():
    """Plots the admin requirements as a line plot."""
    sql = pd.read_csv("admin/admin_sql_seconds.csv")
    it1 = pd.read_csv("admin/admin_it1_seconds.csv")
    it2 = pd.read_csv("admin/admin_it2_seconds.csv")
    sql_aggregated = sql[(sql != sql.min()) & (sql != sql.max())].mean()
    it1_aggregated = it1[(it1 != it1.min()) & (it1 != it1.max())].mean()
    it2_aggregated = it2[(it2 != it2.min()) & (it2 != it2.max())].mean()
    requirements_sql = sql_aggregated.index
    requirements_it1 = it1_aggregated.index
    requirements_it2 = it2_aggregated.index
    plt.figure(figsize=(8, 6))
    plt.plot(requirements_sql, sql_aggregated, marker="o", color="blue")
    plt.plot(requirements_it1, it1_aggregated, marker="^", color="goldenrod")
    plt.plot(requirements_it2, it2_aggregated, marker="s", color="blueviolet")
    plt.axvline(x="A9", linestyle="--", color="black", label="axvline - full height")
    plt.xlabel("Anforderung", color="black")
    plt.ylabel("Laufzeit (in Sekunden)", color="black")
    plt.title("Durchschnittliche Laufzeit der Anforderungen (Administrator)")
    plt.legend(
        ["Transact-SQL", "Iteration 1", "Iteration 2", "Ende der \nIteration 1"], loc=1
    )
    plt.grid()
    plt.ylim(-0.01, 0.41)
    plt.show()


plot_admin_requirements()
```
